```python
import pandas as pd
import numpy as np
import requests 
import pymysql

def get_requests(url, encod=True):
    url = url
    headers = {
        "User-Agent":
            "Mozilla/5.0 (Windows NT 10.0; \
                   WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36"
    }
    response = requests.get(url, headers=headers)
    if encod:
        response.encoding = 'utf-8'
    else:
        response.encoding = 'gbk'
    rs = response.text
    return rs
def sql_table(user_id="0002c7ae0b914ab59f1a15b5fd1d7a56"):
    conn = pymysql.connect(
        host="",
        user="",
        password="",
        db="api_loansuper_x4",
        charset="utf8")
    sql="""    
    SELECT
    a.*
    from
    t_loan_order as a
    LEFT JOIN
    t_loan_order_operate as b on a.loan_id=b.loan_id
    
    WHERE
    a.loan_order_status=5
    and
    b.repay_date>='2021-08-31'
    and
    b.repay_date<='2021-09-03'
    and a.allplatfrom_user_type=1
    and a.extend_times = 0
       """
    df = pd.read_sql(sql, conn)
    conn.close()
    return df
def get_score(loan_id=None,url=None):
    try:
        url=url+loan_id
        score=eval(get_requests(url))["score"]
        return score
    except:
        return -9999

data1=sql_table()

data1=data1.drop_duplicates(subset='loan_id',keep='first')

data=data1

data["star_by"]=data.loan_id

data.index=data.loan_id

import tqdm

data=data[data['extend_times']==0]
```
